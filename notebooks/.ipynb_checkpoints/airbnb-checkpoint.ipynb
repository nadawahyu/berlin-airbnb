```python
def ExtractIO(data, target):
    '''a function to separate input and output columns
    
    params:
        data (dataframe): data of the sample
        target (str): output column
    returns:
        df'''
    
    # input data
    data_input = data.drop(columns = target)
    
    # output data
    data_output = data[target]
    data_output = data_output.replace({',':''}, regex=True).apply(pd.to_numeric,1)
    
    return data_input, data_output
```

```python
# impute missing value with median

from sklearn.impute import SimpleImputer

def ImputerNumerical(data, imputer=None):
    '''imputer for numerical data'''
    
    # check whether imputer exists
    if imputer == None:
        # create imputer
        imputer = SimpleImputer(missing_values = np.nan,
                                strategy = 'median')
        
        # fit imputer
        imputer.fit(data)
        
    # transform data
    data_imputed = pd.DataFrame(imputer.transform(data))
    data_imputed.columns = data.columns
    data_imputed.index = data.index
    
    return data_imputed, imputer
```

#### one hot encoding

```python
from sklearn.preprocessing import OneHotEncoder

def OHETransformer(data, ohe_col = None, encoder = None):
    ''''''
    # check encoder availability
    if encoder == None:
        #create encoder
        encoder = OneHotEncoder(#drop = 'if_binary',
                                handle_unknown = 'ignore')
        # fit
        encoder.fit(data)
        
        # extract ohe cols
        ohe_col = encoder.get_feature_names(data.columns)
        
    # transform data
    data_ohe = encoder.transform(data).toarray()
    data_ohe = pd.DataFrame(data_ohe,
                           columns = ohe_col,
                           index = data.index)
    
    return data_ohe, ohe_col, encoder
```

```python
def ScalerTransformer(data, concat_col, scaler=None):
    ''''''
    if scaler == None:
        scaler = StandardScaler()
        scaler.fit(data)
        
    data_scaled = scaler.transform(data)
    data_scaled = pd.DataFrame(data_scaled, columns = concat_col,
                              index=data.index)
    
    return data_scaled, scaler
```

# transform ds

```python
def TransformTest(data,
                  col_numerical, col_categorical, 
                  ohe_col, concat_col, imputer_num,
                  encoder, scaler
                 ):
    
    # split numerical and categorical data
    data_num, data_cat = SplitCategory(data, col_numerical, col_categorical)
    
    # Handling numerical data
    data_num_imput, _ = ImputerNumerical(data = data_num,
                                         imputer = imputer_num)

    # Handling categorical data
    data_cat_imput = ImputerCategorical(data = data_cat)
    data_cat_ohe, _, _ = OHETransformer(data = data_cat_imput,
                                        ohe_col = ohe_col,
                                        encoder = encoder)
    
    # Concat data
    data_concat = pd.concat([data_num_imput, data_cat_ohe],
                            axis = 1)
    
    # Scaled data
    data_clean, _ = ScalerTransformer(data = data_concat, concat_col = concat_col, scaler = scaler)
    
    return data_clean
```

```python
X_test_clean = TransformTest(X_test, col_numerical, col_categorical, ohe_col, concat_col, imputer_num, encoder, scaler)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
knn = KNeighborsRegressor()
dtr = DecisionTreeRegressor()
lr = LinearRegression()
ridge = Ridge()
lasso = Lasso()
rfr = RandomForestRegressor()
abr = AdaBoostRegressor()
gbr = GradientBoostingRegressor()

model_list = [knn, dtr,           # Region-based 
              lr, ridge, lasso,   # Linear-based
              rfr, abr, gbr]      # Ensemble-based
```

```python
knn_params = {"n_neighbors": [30, 50, 70, 90]}
dtr_params = {"max_depth": [3, 5, 7, 9]}
lr_params = {}
ridge_params = {"alpha": np.logspace(-4, 4, 10)}
lasso_params = {"alpha": np.logspace(-4, 4, 10)}
rfr_params = {"n_estimators": [100, 300, 500]}
abr_params = {"n_estimators": [100, 300, 500]}
gbr_params = {"n_estimators": [100, 300, 500]}

param_list = [knn_params, dtr_params,
              lr_params, ridge_params, lasso_params,
              rfr_params, abr_params, gbr_params]
```

### gridsearch cv

```python
from sklearn.model_selection import GridSearchCV
```

```python
best_param_list = []
best_param_score = []

for i in range(len(model_list)):
    estimator = model_list[i]
    estimator_param = param_list[i]
    
    # Buat CV
    estimator_cv = GridSearchCV(estimator = estimator,
                                param_grid = estimator_param,
                                cv = 5,
                                scoring = "neg_root_mean_squared_error")
    
    # Fit CV
    print(f"start training model: {i+1}")
    estimator_cv.fit(X_train_clean, y_train)
    
    # Print results
    best_param_cv = estimator_cv.best_params_
    best_score_cv = -estimator_cv.best_score_
    best_param_list.append(best_param_cv)
    best_param_score.append(best_score_cv)
    print(f"best params: {best_param_cv}")
    print(f"best scores: {best_score_cv:.4f}")
    print("")
```
